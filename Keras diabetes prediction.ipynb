```python
from tensorflow import keras
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load the data
file_path = (r'C:\Users\User\Documents\MLAIstuffs\pima-indians-diabetes.data.csv')
dataset = loadtxt(file_path, delimiter=',')

# Split the data into training and test sets
x = dataset[:,0:8]
y = dataset[:,8]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Define the hyperparameter search space
space = {
    'num_nodes_1': keras.layers.core.Dropout(0.1),
    'num_nodes_2': keras.layers.core.Dropout(0.1),
    'num_nodes_3': keras.layers.core.Dropout(0.1),
    'activation_1': 'relu',
    'activation_2': 'relu',
    'activation_3': 'sigmoid',
    'num_epochs': 150,
    'batch_size': 10,
    'learning_rate': 0.001
}

# Define the model
def build_model(hp):
    model = Sequential()
    model.add(Dense(hp['num_nodes_1'], input_shape=(8,), activation=hp['activation_1']))
    model.add(Dense(hp['num_nodes_2'], activation=hp['activation_2']))
    model.add(Dense(hp['num_nodes_3'], activation=hp['activation_3']))
    model.compile(loss='binary_crossentropy', optimizer=keras.optimizers.Adam(hp['learning_rate']), metrics=['accuracy'])
    return model

# Define the tuner
tuner = keras.wrappers.scikit_learn.KerasClassifier(build_fn=build_model, epochs=space['num_epochs'], batch_size=space['batch_size'])

# Define the hyperparameter tuning algorithm
tuner.search(x_train, y_train, cv=3, scoring='accuracy')

# Get the best hyperparameters
best_params = tuner.get_best_hyperparameters()[0]

# Build the model with the best hyperparameters
model = build_model(best_params)

# Train the model on the training set
model.fit(x_train, y_train, epochs=best_params['num_epochs'], batch_size=best_params['batch_size'], validation_data=(x_test, y_test))

# Evaluate the model on the test set
y_pred = model.predict_classes(x_test)
accuracy = accuracy_score(y_test, y_pred)
print('Test accuracy:', accuracy)
```
